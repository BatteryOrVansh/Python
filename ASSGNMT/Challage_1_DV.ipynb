````python
import matplotlib.pyplot as plt
import pandas as pd



````

````python
data = {'Date': ['2023-12-01','2023-12-02','2023-12-03','2023-12-04','2023-12-05'],'Temp':[15,18,12,10,17]}

df=pd.DataFrame(data)
print(df)

plt.xlabel('Date')
plt.ylabel('Temp (HIGH)')
plt.title('Analyzing Weather Data')
plt.plot(df.Date,df.Temp)

plt.bar(df.Date,df.Temp,color='black')
````

````python
data = {'Title': ['The Shawshank Redemption', 'The Godfather', 'The Dark Knight', 'The Godfather: Part II', 'Pulp Fiction', 'Lord of the Rings: The Return of the King', 'Schindler\'s List', '12 Angry Men', 'The Lord of the Rings: The Fellowship of the Rings', 'Fight Club'],'Genre': ['Drama','Crime','Action','Crime','Crime','Fantasy','Historical Drama','Drama','Fantasy','Drama']}
df=pd.DataFrame(data)
print(df)



plt.plot(df.Genre,df.Title)
````
