````python
import matplotlib.pyplot as plt
import pandas as pd
````

````python
data = {'Country':['Japan','Swizerland','Singapore','Australia','Iceland'],
        'Life_Expectancy':[84.3,84.2,83.7,82.8,83.3],'GDP':[39522,87169,65870,51408,67004]}
df=pd.DataFrame(data)
````

````python
plt.scatter(df['Life_Expectancy'],df['GDP'])

for i, txt in enumerate(df['Country']):
    plt.text(df['Life_Expectancy'][i], df['GDP'][i], txt, ha='left') # this current loop is from chatgpt and i fully understand this code now!

plt.xlabel('Life Expectancy')
plt.ylabel('GDP per capita (USD)')
plt.title('World Data')
plt.show()
````
