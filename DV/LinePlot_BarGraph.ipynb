## 'LINE PLOT'

# 1. create a list with name tata and put 5 values as per your choice. create a list with day name from monday to friday. create line plot
# 2. create a list of seven months and create another list with the marks of each month respectively. create a line plot and observe the trend.
# 3. we have a friend named Rohit and he's a cricketer. In his last series of 5 matches, he scored 25,0,98,54 and 10. create a visualization of this tabular data

```python
tata=[15,20,30,25,35]
days=["monday", "tuesday", "wednesday", "thursday", "friday"]
plt.plot(days,tata)
plt.xlabel("Days")
plt.ylabel("Prices")
plt.title("STOCK PRICES")
```

```python

month=["jan","feb","mar","apr","may","jun","july"]
marks=[80,67,88,73,77,89,69]
plt.plot(month,marks)
plt.xlabel('DAYS')
plt.ylabel('MONTHS')
plt.title('MONTHLY MARKS')
```

```python
match=['1st','2nd','3rd','4th','5th']
score=[25,0,98,54,10]
plt.plot(match,score,color="green")
plt.xlabel('Matches')
plt.ylabel('Runs')
plt.title("Rohit's 5 innings")
```

## BAR GRAPH
# 1. Create a list of 5 students with their marks in python and create a bar graph to observe who got the highest and the lowest marks.

```python
st=["Harry","Mark","John","Rob","Chris"]
marks=[78,87,98,66,94]
plt.figure(figsize=(5,4))
plt.xlabel("STUDENTS")
plt.ylabel("MARKS")
plt.title("STUDENTS' MARKS")
plt.bar(st,marks,color="black")
```
