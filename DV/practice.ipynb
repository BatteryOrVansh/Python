```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
data={'Days':np.arange(1,31),'Stock_Prices':np.random.randint(100,150,size=30)}
df=pd.DataFrame(data)
df
```

```python
plt.plot(df['Days'],df['Stock_Prices'])
plt.xlabel('DAYS OF MONTH')
plt.ylabel('STOCK PRICES')
plt.title('Monthly Stock Prices Trends')
plt.show()
```

```python
movie_genres=['Comedy','Action','Thriller','Drama']
genre_counts=[20,35,15,25]
plt.bar(movie_genres,genre_counts,color='pink',edgecolor='black')

plt.xlabel('Genre')
plt.ylabel('Number of Movies')
plt.title('Distribution of Movies')
plt.show()
```

```python
data={'customer_ages':np.random.randint(18,65,size=100),
      'purchase_amounts':np.random.randint(100,2000,size=100)}
df=pd.DataFrame(data)
plt.scatter(df['customer_ages'],df['purchase_amounts'])
plt.xlabel('Customer Age')
plt.ylabel('Purchase Amount')
plt.title('Customer Age vs. Purchase Amount')
plt.show()

```
